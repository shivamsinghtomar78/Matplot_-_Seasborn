**Scatter plots**

The scatter plot is a mainstay of statistical visualization .it depicts the joint distribution of two variables using a cloud of points , where each point represent an observation in the dataset.This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship betweeen them

As we will see ,these visualization can be quite helpful because they use simple and easily understood representation of data even on complex data by mapping up to three additional variables using semantics of hue ,size and style.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

Plot a scatter plot between total_bill and tip

```python
#plot here
sns.relplot(x='total_bill',y='tip',kind="scatter",data=data)
```

The hue parameters


```python
sns.relplot(x='total_bill',y='tip',hue='smoker',kind='scatter',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex',kind='scatter',data=data)
```

The style parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style="sex",kind='scatter',data=data)
```

The size parameter:

The continuous and numerical data

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style="sex",size="size",kind='scatter',data=data)
```

**Using the scatterplot function**

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=data)
```

PLotting with catergorical data

Catplot[Figure level Function]


```python
#x-axis ---> categorical data
#y-axis ---->Numerical data
```

```python
plt.style.use("fivethirtyeight")
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue="sex",data=tips)
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
```
