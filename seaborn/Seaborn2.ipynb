**BOXPLOT:**

A boxplot is a standarized way of displaying the distribution of data based on a five number summary ("minimum ,first quartile(Q1),median,third quartile(Q3) and maximum),it can tell you about your outliers and what theire values are .It can also tell you if your data is symmetrical ,how tightly your data is grouped and if and how your data skewed.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
#univariate
sns.boxplot(tips["tip"])
```

```python
#univariate
sns.boxplot(tips["total_bill"])
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="box")
```

```python
sns.catplot(x="day",y="total_bill",hue="sex",data=tips,kind="box")
```

**Violin plots :**

Violin plots are a method of plotting numeric data and can be considered a combination of the box with kernel density plot

Violin plots often used to compare the distribution of a given variable across some categories

```python
sns.violinplot(tips["total_bill"])
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="violin")
```

```python
sns.catplot(x="day",y="total_bill",hue="sex",data=tips,kind="violin")
```

```python
sns.catplot(x="day",y="total_bill",hue="sex",data=tips,split=True,kind="violin")
```

**BarPLot and CountPlot**

```python
sns.catplot(x="smoker",y="total_bill",data=tips,kind="bar")
```

```python
sns.catplot(x="smoker",y="total_bill",hue="sex",data=tips,kind="bar")
```

**HeatMap:**

A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors

```python
flights=sns.load_dataset("flights")
```

```python
flights.head()
```

```python
x=flights.pivot_table(index="year",columns="month",values="passengers")
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt="g")
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt="g",cmap="YlGnBu")
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt="g",cmap="viridis")
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt="g",cmap="summer")
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt="g",cmap="winter")
```

**ClusterMap:**

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,z_score=0,annot=True,metric="correlation")
```

**JointPlot**

Joinplot creates a multi-panel figure that shows both the bivariate(or joint) relationship between two variable along with the univariate (or marginal) distribution of each on separate axes.

we can use JoinPlots in the following 4 manners:

1.scatterplots
2.Hexbin plots
3.Kdeplots
4.Regplots

```python
tips=sns.load_dataset("tips")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

**Pairplots**

Pairplots are used to plot multiple pairwise bivariate distributions in a dataset. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame. By default, it also draws the univariate distribution of each variable on the diagonal Axes.

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue="species")
```

**Distplots**

Distplots are used to do univariate analysis on a continuous/numerical data.

Distplot combines 3 things in one graph:

1.Histogram

2.Kernel Density Estimation (PDF)

3.Rugplot

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic["age"] = titanic["age"].fillna(titanic["age"].mean())
```

Plotting a DistPlot

```python
sns.distplot(titanic["age"])
```

Plotting only the histogram

```python
sns.distplot(titanic["age"],kde=False)
```

PLotting only the Kde

```python
sns.distplot(titanic["age"],hist=False)
```

PLotting only the RugPlot

```python
sns.distplot(titanic["age"],hist=False,rug=True,kde=False)
```
