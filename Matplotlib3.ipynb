```python
data=pd.read_csv("data.csv")
```

```python
lang_counts = {}

for index, row in data.iterrows():
    languages = row['LanguagesWorkedWith'].split(';')
    for language in languages:
        lang_counts[language] = lang_counts.get(language, 0) + 1

languages = list(lang_counts.keys())
counts = list(lang_counts.values())

# Sort languages by count in descending order
sorted_indices = sorted(range(len(counts)), key=lambda k: counts[k], reverse=True)
languages = [languages[i] for i in sorted_indices]
counts = [counts[i] for i in sorted_indices]

# Select top N languages (e.g., top 15)
top_n = 15
languages = languages[:top_n]
counts = counts[:top_n]

plt.figure(figsize=(10, 6))
plt.bar(languages, counts)
plt.xlabel("Languages")
plt.ylabel("Number of Developers")
plt.title("Most Popular Programming Languages")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
